# Importing Table from WikiPedia

### Importing BeautifulSoup

```python
from bs4 import BeautifulSoup
import requests

url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
r = requests.get(url)
soup = BeautifulSoup(r.content)
```

### Seek out table to import

```python
right_table=soup.find('table', class_='wikitable sortable')
```

### Append the Table data to each list

```python
A=[]
B=[]
C=[]

for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:   #change to number of column
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

### Append the Table data to each Column of the Dataframe

```python
import pandas as pd
df=pd.DataFrame(A,columns=['Postal Code'])
df['Borough']=B
df['Neignbourhood']=C
df.head()
```

### Filter out the Borough's and Neigbourhood with a "Not Assigned" value

```python
new_df = df[~df["Borough"].str.contains("Not assigned")]
new_df.head()
```
